```python
import numpy as np
import csv
import time
import itertools
from binary_clt import BinaryCLT

def load_data(filename):
    """Load data from a CSV file."""
    with open(filename, 'r') as file:
        reader = csv.reader(file, delimiter=',')
        return np.array(list(reader)).astype(np.float64)
```
